```python
import spacepy.time as spt
import spacepy.coordinates as spc
import spacepy.irbempy as ib
import numpy as np

t = spt.Ticktock(['2017-02-02T12:00:00'], 'ISO')
```

```python
rad_range_e = np.linspace(1.01,20,41)
lat_range_e = np.linspace(0,  180,41)

energy_e = 1.0 #MeV
energy_p = 10.0 #MeV

AE8 = np.zeros((len(rad_range_e),len(lat_range_e)))

for i in np.arange(0,len(rad_range_e)):
    for j in np.arange(0,len(lat_range_e)):
        y = spc.Coords([[rad_range_e[i]*np.sin(lat_range_e[j]*np.pi/180.0),0.0,
                         rad_range_e[i]*np.cos(lat_range_e[j]*np.pi/180.0) ,0.0]], 'MAG', 'car')
        y.ticks = t
        
        AE8[i,j] = ib.get_AEP8(energy_e, y, model='max', fluxtype='diff', particles='e')
        print("RAD :"+str(rad_range_e[i])+", LAT : "+str(lat_range_e[j])+" AE8 FLUX: "+str(AE8[i,j]))
        
        
rad_range_p = np.linspace(1.01,10,41)
lat_range_p = np.linspace(0,  180,41)


AP8 = np.zeros((len(rad_range_p),len(lat_range_p)))

for i in np.arange(0,len(rad_range_p)):
    for j in np.arange(0,len(lat_range_p)):
        y = spc.Coords([[rad_range_p[i]*np.sin(lat_range_p[j]*np.pi/180.0),0.0,
                         rad_range_p[i]*np.cos(lat_range_p[j]*np.pi/180.0) ,0.0]], 'MAG', 'car')
        y.ticks = t
        
        AP8[i,j] = ib.get_AEP8(energy_p, y, model='max', fluxtype='diff', particles='p')
        print("RAD :"+str(rad_range_p[i])+", LAT : "+str(lat_range_p[j])+" AP8 FLUX: "+str(AP8[i,j]))        
```

```python
AE8[np.isnan(AE8)] = 2.0
AP8[np.isnan(AP8)] = 2.0

XE = np.zeros((len(rad_range_e),len(lat_range_e)))
ZE = np.zeros((len(rad_range_e),len(lat_range_e)))


XP = np.zeros((len(rad_range_p),len(lat_range_p)))
ZP = np.zeros((len(rad_range_p),len(lat_range_p)))


for i in np.arange(0,len(rad_range_e)):
    for j in np.arange(0,len(lat_range_e)):
        
        XE[i,j] = rad_range_e[i]*np.sin(lat_range_e[j]*np.pi/180.0)
        ZE[i,j] = rad_range_e[i]*np.cos(lat_range_e[j]*np.pi/180.0)
        
for i in np.arange(0,len(rad_range_p)):
    for j in np.arange(0,len(lat_range_p)):   
        
        XP[i,j] = -1*rad_range_p[i]*np.sin(lat_range_p[j]*np.pi/180.0)
        ZP[i,j] = rad_range_p[i]*np.cos(lat_range_p[j]*np.pi/180.0)
```

```python
import matplotlib.pyplot as plt
from matplotlib import ticker, cm
from matplotlib.colors import LogNorm

plt.rc('font',family='Times New Roman')
params = {'mathtext.default': 'regular' }
plt.rcParams.update(params)

plt.figure(figsize=(19,8))
ax = plt.gca()
norme=LogNorm(vmin=AE8.min(), vmax=AE8.max())

axe = plt.contourf(XE, ZE, AE8, locator=ticker.LogLocator(subs=range(1,5)), norm=norme, cmap='jet')
cb1 = plt.colorbar(axe)
cb1.set_label('Electron flux \n'+r'$ E > 1$'+' MeV, '+r'$cm^{-2}s^{-1}$', labelpad=-27, y=1.15, rotation=0, fontsize=18)
cb1.ax.tick_params(axis='y', direction='in')
cb1.ax.tick_params(labelsize=18)

normp=LogNorm(vmin=AP8.min(), vmax=AP8.max())

axp = plt.contourf(XP, ZP, AP8, locator=ticker.LogLocator(subs=range(1,4)), norm=normp, cmap='jet')
cb2 = plt.colorbar(axp, ax=[ax], location='left')
cb2.set_label('Proton flux \n'+r'$ E > 10$'+' MeV, '+r'$cm^{-2}s^{-1}$', labelpad=-27, y=1.05, rotation=0, fontsize=18)
cb2.ax.tick_params(axis='y', direction='in')
cb2.ax.tick_params(labelsize=18)

ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')

ax.set_xlabel("X, "+r'$R_E$' ,fontsize=18)
ax.set_ylabel("Z, "+r'$R_E$' ,fontsize=18)

ax.tick_params(axis='x',labelsize=18)
ax.tick_params(axis='y',labelsize=18)

ax.tick_params('both', length=5, width=2, which='major', color='w')

r0 = 1.0

theta = np.linspace(0,2*np.pi,101)

x_arr = r0*np.cos(theta)
y_arr = r0*np.sin(theta)

plt.plot(x_arr, y_arr, 'k-', linewidth=3.0)

plt.axvline(x=0.0, linestyle='solid', color='w', linewidth=2.0)

plt.text(2,4,'Electrons', ha='center', va='center', color='white', fontsize=18)
plt.text(-2,4,'Protons', ha='center', va='center', color='white', fontsize=18)

plt.title('Geomagentic coordinates', fontsize=18)

plt.xlim([-5,10])
plt.ylim([-5,5])

plt.savefig('../new_plots/rad-belt-e-1-p-10.png',bbox_inches='tight')
plt.savefig('../new_plots/rad-belt-e-1-p-10.pdf', dpi=300,bbox_inches='tight')
plt.savefig('../new_plots/rad-belt-e-1-p-10.eps', dpi=300,bbox_inches='tight')
```

```python
rad_range_e = np.linspace(1.01,10,101)
ene_range_e = np.array([0.1, 0.2, 0.4, 0.8, 1.0, 2.0, 4.0])


AE8 = np.zeros((len(rad_range_e),len(ene_range_e)))

for i in np.arange(0,len(rad_range_e)):
    for j in np.arange(0,len(ene_range_e)):
        y = spc.Coords([[rad_range_e[i]*np.sin(90.0*np.pi/180.0),0.0,
                         rad_range_e[i]*np.cos(90.0*np.pi/180.0) ,0.0]], 'MAG', 'car')
        y.ticks = t
        
        AE8[i,j] = ib.get_AEP8(ene_range_e[j], y, model='max', fluxtype='diff', particles='e')
        print("RAD :"+str(rad_range_e[i])+", ENE : "+str(ene_range_e[j])+" AE8 FLUX: "+str(AE8[i,j]))
        
        
rad_range_p = np.linspace(1.01,5,101)
ene_range_p = np.array([0.1, 1.0, 10.0, 50.0, 100.0, 400.0])


AP8 = np.zeros((len(rad_range_p),len(ene_range_p)))

for i in np.arange(0,len(rad_range_p)):
    for j in np.arange(0,len(ene_range_p)):
        y = spc.Coords([[rad_range_p[i]*np.sin(90.0*np.pi/180.0),0.0,
                         rad_range_p[i]*np.cos(90.0*np.pi/180.0) ,0.0]], 'MAG', 'car')
        y.ticks = t
        
        AP8[i,j] = ib.get_AEP8(ene_range_p[j], y, model='max', fluxtype='diff', particles='p')
        print("RAD :"+str(rad_range_p[i])+", ENE : "+str(ene_range_p[j])+" AP8 FLUX: "+str(AP8[i,j]))        
```

```python
AE8[np.isnan(AE8)] = 1.0
AP8[np.isnan(AP8)] = 1.0
```
