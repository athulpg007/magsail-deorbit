```python
import magpylib as magpy
import numpy as np
from matplotlib import pyplot as plt
```

```python
# set font size and define figures
plt.rcParams.update({'font.size': 18})
```

```python
s1 = magpy.source.current.Line(curr=1E2, vertices=[(0,-3140,0),(0,3140,0)])
s2 = magpy.source.current.Circular(curr=1E2, dim=2000) 
```

```python
x_arr  = np.linspace(10.0,1000000,10001)

posis2 = np.array([(x,0,0) for x in x_arr])

B1            = np.array([s1.getB(p) for p in posis2])
B2            = np.array([s2.getB(p) for p in posis2])
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

plt.figure(figsize=(8,8))
plt.rc('font',family='Times New Roman')
params = {'mathtext.default': 'regular' }
plt.rcParams.update(params)

plt.plot(x_arr/1E3, abs(B1[:,2])/1E3, linewidth=5.0, color='g', label='Line L = 6.28 m' )
plt.plot(x_arr/1E3, abs(B2[:,2])/1E3, linewidth=5.0, color='b', label='Wire D = 2.00 m' )

x_arr_1 = np.linspace(0.01,100,101)
y_arr_1 = 0.00002/x_arr_1

x_arr_2 = np.linspace(0.1,1000,101)
y_arr_2 = 0.00005/x_arr_2**2

x_arr_3 = np.linspace(0.1,1000,101)
y_arr_3 = 0.000040/x_arr_3**3



plt.xlabel('X, m',fontsize=18)
plt.ylabel('B, T', fontsize=18)

#plt.plot(x_arr_1, y_arr_1, linewidth=1.0, linestyle='dashed', color='k', label='1/r' )
plt.plot(x_arr_2, y_arr_2, linewidth=1.0, linestyle='dashdot', color='g', label=r'$1/r^2$' )
plt.plot(x_arr_3, y_arr_3, linewidth=1.0, linestyle='dashdot', color='b', label=r'$1/r^3$' )

#plt.axhline(y=2.7E-4, linestyle='dashed', linewidth=2.0, color='red')


#plt.text(5.0, 150.0, 'P = 100 W\nwire_radius = 2.2 mm\ns/c mass = 10 kg', fontsize=20)
ax=plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')

ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=18)
ax.tick_params(axis='y',labelsize=18)

ax.set_xscale('log')
ax.set_yscale('log')

ax.xaxis.set_tick_params(direction='in', which='both')
ax.yaxis.set_tick_params(direction='in', which='both')

ax.xaxis.set_tick_params(width=2, length=8)
ax.yaxis.set_tick_params(width=2, length=8)

ax.xaxis.set_tick_params(width=1, length=4, which='minor')
ax.yaxis.set_tick_params(width=1, length=4, which='minor')


ax.xaxis.grid(which='major', color='k', linestyle='solid', linewidth=0.2)
ax.xaxis.grid(which='minor', color='k', linestyle='solid', linewidth=0.1)

ax.yaxis.grid(which='major', color='k', linestyle='solid', linewidth=0.2)
ax.yaxis.grid(which='minor', color='k', linestyle='solid', linewidth=0.1)

plt.legend(loc='upper right', fontsize=18)

plt.savefig('../new_plots/line-vs-loop.png',bbox_inches='tight')
plt.savefig('../new_plots/line-vs-loop.pdf', dpi=300,bbox_inches='tight')
plt.savefig('../new_plots/line-vs-loop.eps', dpi=300,bbox_inches='tight')

plt.show()
```
