```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
i = 1

den = np.load('1den_'+str(i).zfill(4)+'.npy')

x = np.linspace(-0.1,0.1,21)
z = np.linspace(0.0, 0.4, 41)

zgrid, xgrid = np.meshgrid(z,x)

plt.figure(figsize=(9.5*1.4,4*1.4))
plt.rc('font',family='Times New Roman')
plt.contourf(zgrid,xgrid,den[:,10,:]/1E10, levels=21)
cbar=plt.colorbar()
cbar.ax.tick_params(labelsize=18)
cbar.set_label(r'$N \times 10^{10}$', labelpad=-27, y=1.10, rotation=0, fontsize=18)

x0 = 0
z0 = 0.15

r0 = 0.05

theta =np.linspace(0,2*np.pi,101)

x1 = x0 + r0*np.cos(theta)
z1 = z0 + r0*np.sin(theta)

plt.plot(z1,x1,'w-')

ax = plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
plt.tick_params(direction='in')
plt.tick_params(axis='x',labelsize=18)
plt.tick_params(axis='y',labelsize=18)

plt.xlabel('Z, m',fontsize=18)
plt.ylabel("X, m",fontsize=18)

plt.title("t = "+str(round(i*1E-1,3))+" "+r'$\mu s$', fontsize=18)
plt.savefig('1den_'+str(i).zfill(4)+'.png', bbox_inches='tight', dpi=200)

plt.show()
```
