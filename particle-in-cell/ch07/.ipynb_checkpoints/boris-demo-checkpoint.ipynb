```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class World:
    
    def __init__(self):
        
        self.EPS_0 = 8.85418782e-12    # epsilon_0
        self.QE    = 1.602176565e-19;  # proton charge
        self.AMU   = 1.660538921e-27   # amu
        self.ME    = 9.10938215e-31;   # mass of electron    
        self.K     = 1.380648e-23;     # Boltzmann's const.
        self.EvToK = self.QE/self.K;   # 1 eV = q/k
```

```python
class Particle:
    
    def __init__(self,q, m, pos, vel):
        
        self.q   = q
        self.m   = m
        self.pos = pos
        self.vel = vel
        
    def computeLarmorRadius(self,vt,E,B):
        
        return (self.m*vt)/(abs(self.q)*np.linalg.norm(B))
        
    def updateVelocityDirect(self, dt, E, B):
        
        acc      = (self.q/self.m)*(E + np.cross(self.vel, B))
        self.vel = self.vel + acc*dt
    
    def updateVelocityBoris(self, dt, E, B):
        
        v_minus = self.vel + (self.q/self.m)*E*dt/2
        t       = (self.q/self.m)*B*dt/2
        v_prime = v_minus + np.cross(v_minus, t)
        s       = 2.0*t/(1+np.dot(t,t))
        v_plus  = v_minus + np.cross(v_prime, s)
        self.vel= v_plus + (self.q/self.m)*E*dt/2      
```

```python
world = World()

E = np.array([0,0,0])
B = np.array([0,0,0.01])

part1 = Particle(-world.QE, world.ME, np.array([0,0,0]), np.array([0,0,1.0]))
RL    = part1.computeLarmorRadius(1E3, E, B)
part1.pos = np.array([RL,0,0])
part1.vel = np.array([0, 1E3, 0])

part2 = Particle(-world.QE, world.ME, np.array([0,0,0]), np.array([0,0,1.0]))
RL    = part2.computeLarmorRadius(1E3, E, B)
part2.pos = np.array([RL,0,0])
part2.vel = np.array([0, 1E3, 0])

dt = 5E-11

#rewind velocity per leapfrog
part1.updateVelocityDirect(-0.5*dt,E,B)
part2.updateVelocityBoris (-0.5*dt,E,B)

x_arr1 = np.zeros(300)
y_arr1 = np.zeros(300)

x_arr2 = np.zeros(300)
y_arr2 = np.zeros(300)



for t in range(300):
    
    x_arr1[t] = part1.pos[0]
    y_arr1[t] = part1.pos[1]
    
    x_arr2[t] = part2.pos[0]
    y_arr2[t] = part2.pos[1]
    
    part1.updateVelocityDirect(dt, E, B)
    part2.updateVelocityBoris (dt, E, B)
    
    part1.pos += part1.vel*dt
    part2.pos += part2.vel*dt
    
plt.figure(figsize=(8,8))
plt.rc('font',family='Times New Roman')

plt.plot(x_arr1/1E-6,y_arr1/1E-6, 'k--', lw=2.0, label='Direct')
plt.plot(x_arr2/1E-6,y_arr2/1E-6, 'k-',lw=2.0, label='Boris')

#plt.ticklabel_format(axis='both',style='sci', scilimits=(0,0))

ax = plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
plt.tick_params(direction='in')
plt.tick_params(axis='x',labelsize=18)
plt.tick_params(axis='y',labelsize=18)

plt.legend(loc='upper right', fontsize=18)

tx = ax.xaxis.get_offset_text()
tx.set_fontsize(18)

plt.xlabel('X / 1E-6, m',fontsize=18)
plt.ylabel("Y / 1E-6, m",fontsize=18)

```
