```python
import numpy as np
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
import random
%matplotlib notebook


fig = pyplot.figure(figsize=(9,9))
ax = Axes3D(fig)

x_arr1 = np.loadtxt('species_1_x_01500.txt')
y_arr1 = np.loadtxt('species_1_y_01500.txt')
z_arr1 = np.loadtxt('species_1_z_01500.txt')

x_arr2 = np.loadtxt('species_2_x_01500.txt')
y_arr2 = np.loadtxt('species_2_y_01500.txt')
z_arr2 = np.loadtxt('species_2_z_01500.txt')

ax.view_init(elev=-90., azim=0)
ax.scatter(x_arr1, y_arr1, z_arr1, s=3.5, alpha=0.05)
ax.scatter(x_arr2, y_arr2, z_arr2, s=5)
pyplot.show()
```
