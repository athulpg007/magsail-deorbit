```python
from world import World
from species import Species
import numpy as np
```

```python
world=World(21,21,21)
world.setExtents(-0.1, -0.1, 0.0 , 0.1, 0.1, 0.2)

species1 = Species("O+", 16*world.AMU, world.QE)
species2 = Species("e-", world.ME, -1.0*world.QE)

#species1.loadParticlesBox(world.x0, world.xm, 1E11, 800)
#species2.loadParticlesBox(world.x0, world.xc, 1E11, 100)
```

```python
x_arr1 = np.zeros(800)
y_arr1 = np.zeros(800)
z_arr1 = np.zeros(800)

x_arr2 = np.zeros(100)
y_arr2 = np.zeros(100)
z_arr2 = np.zeros(100)

for i in np.arange(0,800):
    pos = species1.particleList[i].pos
    x_arr1[i] = pos[0]
    y_arr1[i] = pos[1]
    z_arr1[i] = pos[2]
    
for i in np.arange(0,100):
    pos = species2.particleList[i].pos
    x_arr2[i] = pos[0]
    y_arr2[i] = pos[1]
    z_arr2[i] = pos[2]
```

```python
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
import random
%matplotlib notebook


fig = pyplot.figure(figsize=(9,9))
ax = Axes3D(fig)

 

ax.scatter(x_arr1, y_arr1, z_arr1, 'ro')
ax.scatter(x_arr2, y_arr2, z_arr2, 'ko')
pyplot.show()
```
