```python
#!/usr/bin/env python
"""
Example:
Poker Flat Research Range altitude profile
"""
import numpy as np
from datetime import datetime
from matplotlib.pyplot import show

import msise00.plots as msplots
import msise00

glat = 65.1
glon = -147.5
alt_km = np.arange(0.0, 1000, 20.0)
time = datetime(2013, 12, 13, 10, 0, 0)

atmos = msise00.run(time, alt_km, glat, glon)
```

```python
atmos
```

```python
atmos.alt_km
```

```python
import matplotlib.pyplot as plt
plt.plot(atmos.alt_km,atmos.Total[0,:,0,0])
ax=plt.gca()
ax.set_yscale('log')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(atmos.alt_km,atmos.Tn[0,:,0,0])
ax=plt.gca()
plt.show()
```
