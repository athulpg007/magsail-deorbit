# Plot the shape of the ideal dipole magnetopause in the meridian and equatorial planes

## Meridian plane

\begin{equation}
\rho \cos{\theta} = \frac{1.89\rho^3}{1+\rho^3}
\end{equation}

```python
theta_circ = np.linspace(1.0/3.0, np.pi/2, 101)

K = 3.0/(2**(2.0/3.0))


x_arr_circ = 1.0*np.sin(theta_circ)
y_arr_circ = 1.0*np.cos(theta_circ)

rho_arc1 = np.linspace(1.00 ,  1.26, 101)
rho_arc2 = np.linspace(1.27 ,  10.0, 101)

theta2_arr = np.arccos(K*rho_arc1**2/(1+rho_arc1**3))
theta3_arr = -1.0*np.arccos(K*rho_arc2**2/(1+rho_arc2**3))

x_arc1  = rho_arc1*np.sin(theta2_arr)
y_arc1  = rho_arc1*np.cos(theta2_arr)

x_arc2  = rho_arc2*np.sin(theta3_arr)
y_arc2  = rho_arc2*np.cos(theta3_arr)
```
