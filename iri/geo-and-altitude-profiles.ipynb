```python


a=0
b=0

lat = np.linspace(-90,90,20)
lon = np.linspace(-180,180,20)

nO = np.zeros((len(lat),len(lon) ))
nH = np.zeros((len(lat),len(lon) ))

time_start_stop = (datetime(2020, 6, 12, 0, 0, 0), datetime(2020, 6, 13))
time_step = timedelta(days=1)
alt_km_range = (800, 900, 101.)


for i in range(0,len(lat)):
    for j in range(0,len(lon)):
        sim = iri.timeprofile(time_start_stop, time_step, alt_km_range, lat[i], lon[j])
        nO[i,j] = float(sim['nO+'][0])
        nH[i,j] = float(sim['nH+'][0])
    print("LAT = "+str(lat[i])+" done.")
        
```
